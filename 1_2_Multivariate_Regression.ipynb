# Multivariate Regression

We can expand the 'univariate' Linear Regression example with multiple features. Suppose that we use not one but two 'features' as input to predict the median house price, a bivariate linear regression model looks like this:

$$\hat{y} = a\cdot x_1 + b\cdot x_2 + c$$

Where we have to find the optimal values for $a$, $b$ and $c$ to optimally predict the median sales price $y$. We can apply the same workflow, the only change is that we include two features and the target variable in the dataset.

```python
from pipetorch import DFrame
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df = DFrame.read_from_kaggle('fedesoriano/the-boston-houseprice-data')
```

```python
# Use RM and LSTAT as features and PRICE as target variable.
df = df[['RM', 'LSTAT', 'MEDV']]
```

```python
# plot a colored scatter graph of the data (see lecture notebook)
df.train.scatter2d_color(s=10)
```

The plot indicates that bot variables may contribute to a predictive model

# Model

```python
# instantiate a linear regression model
model = LinearRegression()
```

# Train

```python
# fit the model on the training set
model.fit(df.train_X, df.train_y)
```

# Evaluate

For this dataset, there appear to be too many datapoints to use different colors for inner and outer circles to indicate how well the predictions match the correct outcome (we could do that with a smaller subset though). Try to compute the difference between the predicted values and the correct outcomes on the dataset and plot that in a colored scatter plot.

```python
difference = model.predict(df.train_X) - df.train_y
df.train.scatter2d_color(c = difference, s=5)
```
