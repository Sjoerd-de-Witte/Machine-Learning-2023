# Support Vector Machines

SVMs are based on the idea of finding a hyperplane that best divides a dataset into two classes. The 'optimal' hyperplane is found by identifying the Support Vectors, i.e. data points that are closest to this hyperplane, and by finding the hyperlane that maximizes to distance (or margin) to every one of these Support Vectors.

Like with previous models, the only thing that really changes in the workflow is the model that you instantiate.

Note: sklearn actually has a few different implementations of SVM classifiers: SVC and LinearSVC. The models are the same, the difference is that they use different algorithms to find the optimum. At this point it does not matter much which one you use, but you should of course instantiate the one that you import.

# Data

```python
from pipetorch import DFrame
from sklearn.svm import LinearSVC
```

```python
df = DFrame.read_from_kaggle('uciml/red-wine-quality-cortez-et-al-2009')
df['quality'] = df.quality > 5
df = df[['pH', 'alcohol', 'quality']]
```

```python
df.train.scatter2d_class(s=5, noise=0.1)
```

# Model


```python
# instantiate an SVC model
model = LinearSVC()
```

# Train

```python
# fit the model on the training set
model.fit(df.train_X, df.train_y)
```

# Evaluate

```python
# plot the dataset and the decision boundary
df.train.scatter2d_class()
df.train.plot_boundary(model.predict)
```
