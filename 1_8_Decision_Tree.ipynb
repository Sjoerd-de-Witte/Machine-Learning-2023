# Decision Tree

Decision trees iteratively split the population according to decision rules, e.g. gender, income, to identify groups that ideally all belong to the same class, e.g. political preference. Decision trees have a hyperparameter `max_depth` that controls the complexity of the model.

Like with previous models, the only thing that really changes in the workflow is the model that you instantiate.

# Data

```python
from pipetorch import DFrame
from sklearn.tree import DecisionTreeClassifier
```

```python
df = DFrame.read_from_kaggle('uciml/red-wine-quality-cortez-et-al-2009')
df['class'] = df.quality > 5
df = df[['pH', 'alcohol', 'class']]
```

```python
df.train.scatter2d_class(s=5, noise=0.1)
```

# Model

The difference in code is just choosing a different model. You can use the decision_tree model and can set the max_depth parameter to allow the model to use more decision rules and thus create a more complex model.

```python
# Instantiate a DecisionTreeClassifier with max_depth=4
model = DecisionTreeClassifier(max_depth=4)
```

# Train

```python
# fit the model on the training set
model.fit(df.train_X, df.train_y)
```

# Evaluate

```python
# plot the dataset and the decision boundary
df.train.scatter2d_class()
df.train.plot_boundary(model.predict)
```

```python
# Try to rerun the experiment, what setting of max_depth gives the best results?
from matplotlib import pyplot as plt
for x in range(1,7):
  plt.figure()
  model = DecisionTreeClassifier(max_depth=x)
  model.fit(df.train_X, df.train_y)
  df.train.scatter2d_class()
  df.train.plot_boundary(model.predict)
```
